                        **  Python Basics Questions**

1. What is Python, and why is it popular?

Popular because:

Easy to learn and write.

Huge standard library.

2.What is an interpreter in Python?
An interpreter reads and executes code line by line. Python uses an interpreter (like CPython) to convert Python code to machine-readable form at runtime.

3.What are pre-defined keywords in Python?

Pre-defined keywords are reserved words that have special meaning in Python and cannot be used as identifiers (variable names).
Examples: if, else, while, class, def, return, etc

4.Can keywords be used as variable names?
No, keywords cannot be used as variable names. Doing so will raise a SyntaxError.

5.What is mutability in Python?

* Mutability refers to whether an object can be changed after creation.

* Mutable objects: can be changed (e.g., list, dict, set)

* Immutable objects: cannot be changed (e.g., int, float, str, tuple).

6.Why are lists mutable, but tuples are immutable?
* Lists are designed to allow item updates, additions, and deletions, which makes them mutable.
* Tuples are meant for fixed data, so they don’t allow changes, making them immutable.

7.What is the difference between “==” and “is” operators in Python?
* ==: checks if values are equal.

* is: checks if both refer to the same object in memory.

Example:

a = [1, 2]
b = [1, 2]
a == b  # True
a is b  # False

8 What are logical operators in Python?
Logical operators are used to combine conditional statements:
and: True if both operands are True.
or: True if at least one is True.
not: Reverses the result.

9. What is type casting in Python?
Type casting is the process of converting one data type to another.
Example: int("10"), str(25), float("5.6")

10. What is the difference between implicit and explicit type casting?
Implicit casting: Done automatically by Python (e.g., int to float)
x = 5
y = 2.0
z = x + y  # x automatically becomes float
x = "10"
y = int(x)  # Manually converting string to int

11.What is the purpose of conditional statements in Python?
Conditional statements are used to execute different blocks of code depending on certain conditions (if, elif, else).


12.How does the elif statement work?
elif stands for "else if". It lets you check multiple conditions after an initial if.
Only the first True block is executed.
x = 20
if x < 10:
    print("Less than 10")
elif x < 30:
    print("Less than 30")  # Executes this
else:
    print("30 or more")

13.What is the difference between for and while loops

for: used when you know how many times to loop (e.g., looping over a list or range).

while: used when you don't know how many times beforehand — it runs until a condition is False.

4.Write a program that checks if a given word is a Python keyword.


```python
import keyword

word = input("Enter a word to check: ")

if keyword.iskeyword(word):
    print(f"'{word}' is a Python keyword.")
else:
    print(f"'{word}' is not a Python keyword.")
```

5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently
for each.



```python
# Create a list and a tuple
my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

# Modify an element in the list (mutable)
print("Original list:", my_list)
my_list[1] = 99
print("Modified list:", my_list)

# Attempt to modify an element in the tuple (immutable)
print("Original tuple:", my_tuple)
try:
    my_tuple[1] = 99
except TypeError as e:
    print("Error when modifying tuple:", e)
```

6.Write a function to demonstrate the behavior of mutable and immutable arguments.


```python
def modify_list(my_list):
    my_list.append(100)
    print("Inside modify_list:", my_list)

# Function to modify an immutable object (integer)
def modify_integer(num):
    num += 10
    print("Inside modify_integer:", num)

# Original values
my_list = [1, 2, 3]
my_number = 5

# Function calls
modify_list(my_list)
modify_integer(my_number)
```

9. Write a Python program to convert user input from string to integer, float, and boolean types.

```python

user_input = input("Enter a value: ")

# Convert to integer
try:
    int_value = int(user_input)
    print("Integer:", int_value)
except ValueError:
    print("Cannot convert to integer")

# Convert to float
try:
    float_value = float(user_input)
    print("Float:", float_value)
except ValueError:
    print("Cannot convert to float")

# Convert to boolean
bool_value = bool(user_input)
print("Boolean:", bool_value)
```

10. Write code to demonstrate type casting with list elements

```python
str_list = ["10", "20", "30", "40"]

# Convert each element to integer using map()
int_list = list(map(int, str_list))

print("Original list (strings):", str_list)
print("Converted list (integers):", int_list)
```

13. Write a Python program to find the sum of all even numbers between 1 and 50.

```python
even_sum = 0

for i in range(2, 51, 2):  # Start from 2, go till 50 with step 2
    even_sum += i

print("Sum of all even numbers between 1 and 50 is:", even_sum)
```

14. Write a program to reverse a string using a while loop.

```python
text = input("Enter a string: ")
reversed_text = ""
index = len(text) - 1

while index >= 0:
    reversed_text += text[index]
    index -= 1

print("Reversed string:", reversed_text)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
num = int(input("Enter a non-negative integer: "))

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    factorial = 1
    i = 1

    while i <= num:
        factorial *= i
        i += 1

    print(f"The factorial of {num} is: {factorial}")
```
